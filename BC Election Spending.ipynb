## Donations to candidates, Jan. 1 - Aug. 8, 2017

Scanning in data on the contributions to specific candidates and seeing what the properties of the data are.

```python
import pandas as pd
```

```python
candidates = pd.read_csv("C:\Users\CARMANT\Documents\Raw Data files\CBC\BC Election\Election_Spending_Candidates_RAW.csv")
```

```python
candidates.head()
```

```python
candidates.info()
```

Here are the candidates who received the most money and who donated it. I think the "FILER_NAME" field is usually the candidate. Important to note, this does not include information for the following candidates, because they either didn't file on time or were granted an extension by Elections BC due to extenuating circumstances: Steve Darling (Burnaby-Lougheed, Liberal), Garrison Duke (Burnaby Edmonds, Liberal), Scott Hamilton (Delta North, Liberal), Steve Housser (Cowichan Valley, Liberal), Linda Reimer (Port Moody-Coquitlam, Liberal), Merv Ritchie (Land, Air, Water), Roy Sakata (Richmond Steveston, Greens), Garry Begg (Surrey-Guildford, NDP), Saira Aujla (Surrey-Green Timbers, Greens), Harry Lali (Fraser-Nicola, NDP), Donna Barnett (Cariboo-Chilcotin, Liberal), Doug Donaldson (Stikine, NDP), Rita Helen Giesbrecht (Cariboo-Chilcotin, Green), Spencer Chandra Herbert (Vancouver-West End, NDP) and Colleen Ross (Boundary-Similkameen, NDP)


```python
candidates.groupby(["CONTRIBUTOR_NAME", "AFFILIATION", "ELECTORAL_DISTRICT", "FILER_NAME"]).AMOUNT.sum().reset_index().sort_values("AMOUNT", ascending=False)
```

```python
liberals = candidates[candidates.AFFILIATION == "BC LIBERAL PARTY"]
```

```python
ndp = candidates[candidates.AFFILIATION == "BC NDP"]
```

```python
greens = candidates[candidates.AFFILIATION == "BC GREEN PARTY"]
```

Here are the amounts given to specific candidates for each of the parties. As you can see, NDP candidates received the most.

```python
liberals.AMOUNT.sum()
```

```python
ndp.AMOUNT.sum()
```

```python
greens.AMOUNT.sum()
```

Here are the top contributors to Liberal candidates.

```python
liberals.groupby(["CONTRIBUTOR_NAME", "ELECTORAL_DISTRICT", "FILER_NAME"]).AMOUNT.sum().reset_index().sort_values("AMOUNT", ascending=False)
```

Here are the amounts given to NDP candidates.

```python
ndp.groupby(["CONTRIBUTOR_NAME", "ELECTORAL_DISTRICT", "FILER_NAME"]).AMOUNT.sum().reset_index().sort_values("AMOUNT", ascending=False)
```

Here are the amounts given to Green candidates.

```python
greens.groupby(["CONTRIBUTOR_NAME", "ELECTORAL_DISTRICT", "FILER_NAME"]).AMOUNT.sum().reset_index().sort_values("AMOUNT", ascending=False)
```
